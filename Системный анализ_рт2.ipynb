```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 10)
transformed = pca.fit_transform(df)
transformed
```

```python
print("Координата первого объекта относительно первой главной компоненты:", transformed[0,0])
```

```python
print("Координата первого объекта относительно второй главной компоненты:", transformed[0,1])
```

```python
pca = PCA(n_components = 2)
transformed = pca.fit_transform(df)
transformed

explained_variance_ratio = pca.explained_variance_ratio_

print("Доля объясненной дисперсии первыми двумя главными компонентами:", sum(explained_variance_ratio))
```

```python
pca = PCA(n_components=df.shape[1])
principal_components = pca.fit_transform(df)

explained_variance_ratio = pca.explained_variance_ratio_

cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

num_components = np.argmax(cumulative_variance_ratio > 0.85) + 1

print("Минимальное количество главных компонент для достижения доли объясненной дисперсии более 0.85:", num_components)
```
