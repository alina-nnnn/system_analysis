```python
df = pd.DataFrame({'xi': [1, 2, 3, 4, 5, 6, 7, 8, 9],
                  'a1': ['T','T','T','F','F','F','F','T','F'],
                  'a2': ['T','T','F','F','T','T','F','F','T'],
                  'a3': [5.0, 7.0, 8.0, 3.0, 7.0, 4.0, 5.0, 6.0, 1.0],
                  'class' : ['Y','Y','N','Y','N','N','N','Y','N',]})

df = df.set_index('xi')
df
```

```python
inputt = input(' :').split()
inputt = list(inputt)
inputt
```

```python
a1 = df['a1'].tolist()
a2 = df['a2'].tolist()
a3 = df['a3'].tolist()
class1 = df['class'].tolist()
print(a1)
print(a2)
print(a3)
print(class1)
```

```python
pyn = len(class1)
p_y = class1.count('Y')
p_n = class1.count('N')
p_1_y = len(df[(df['a1'] == inputt[0]) & (df['class'] == 'Y')]) / p_y
p_2_y = len(df[(df['a2'] == inputt[1]) & (df['class'] == 'Y')]) / p_y
p_3_y = len(df[(df['a3'] == inputt[2]) & (df['class'] == 'Y')]) / p_y
p_1_n = 1 - p_1_y 
p_2_n = 1 - p_2_y
p_3_n = 1 - p_3_y
```

```python
py = (p_y / pyn) * p_1_y * p_2_y * p_3_y
pn = (p_n / pyn) * p_1_n * p_2_n * p_3_n
if py > pn:
    print(':  Y')
else:
    print(':  N')    
print('  ', inputt, '  Y :', py)
print('  ', inputt, '  N :', pn)
```
